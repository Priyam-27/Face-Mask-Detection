### Import libraries

```python
import cv2
import numpy as np
from keras.models import load_model
```

```python
loaded_model=load_model('FaceMask.h5')
```

```python
loaded_model.summary()
```

```python
def predict_percentage(img):
    r_image=cv2.resize(img, (100,100))
    Img=r_image.reshape(1,100,100,3)
    for i in loaded_model.predict_proba(Img):
        for j in i:
            return(f'{(1-j)*100}%')
```

```python
def predict_class(img):
    r_image=cv2.resize(img, (100,100))
    Img=r_image.reshape(1,100,100,3)
    
    if loaded_model.predict(Img) >= 0.5:
        return 'Without Mask'
    else:
        return 'With Mask'
```

```python
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
def detect_face(img):
    gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces=face_classifier.detectMultiScale(gray, 1.1,5)
    return faces
```

```python
def classify_image(img):
    if detect_face(img) is ():
        print('No face detected!!!')
    else:
        if predict_class(img) == 'Without Mask':
            for (x,y,w,h) in detect_face(img):
                cv2.rectangle(img, (x,y), (x+w,y+h), (0,0,255), 2)
                cv2.putText(img, 'Without Mask', (x, y), cv2.FONT_HERSHEY_PLAIN, 1.5,(0,0,255),2)
                cv2.putText(img, predict_percentage(img), (x+w-40, y+h+20), cv2.FONT_HERSHEY_PLAIN, 1.5,(0,0,255),2)
        elif predict_class(img) == 'With Mask':
            for (x,y,w,h) in detect_face(img):
                cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)
                cv2.putText(img, 'With Mask', (x, y), cv2.FONT_HERSHEY_PLAIN, 1.5,(0,255,0),2)
                cv2.putText(img, predict_percentage(img), (x+w-70, y+h+20), cv2.FONT_HERSHEY_PLAIN, 1.5,(0,255,0),2)
    return img
        
```

```python
img=cv2.imread('without_mask.jpg')
# img=cv2.resize(img, (640,640))
cv2.imshow('Image', img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
cv2.imshow('Capture',classify_image(img) )
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

### Reading web cam

```python
cap = cv2.VideoCapture(0)
cap.set(3,640)
cap.set(4,640)
cap.set(10,1000)

if (cap.isOpened() == False):
    print('Error in opening WebCam')
else:
    while cap.isOpened:
        ret,img= cap.read()
        if ret==True:
#             IMG=classify_image(img)
            cv2.imshow('Capture', classify_image(img))
            if cv2.waitKey(25) & 0xFF == ord('q'):
                break
        else:
            break
cap.release()
cv2.destroyAllWindows()
```
